```python
import cv2
import os
import pyttsx3
from moviepy.editor import VideoFileClip, AudioFileClip

# Prompt the user to enter the text
text = input("Enter the text to convert to speech: ")

# Initialize pyttsx3 and generate the voice
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)  # Selecting the first index for a male voice
output_file = "output.mp3"
engine.save_to_file(text, output_file)
engine.runAndWait()

# Get the duration of the generated audio
audio_duration = AudioFileClip(output_file).duration

# Path to the folder containing the images
image_folder = input("Enter the path to the folder containing the images: ")

# Retrieve the list of image file names in the folder with supported extensions
supported_extensions = ['.jpg', '.jpeg', '.png']
images = [img for img in os.listdir(image_folder) if os.path.splitext(img)[1].lower() in supported_extensions]

# Sort the image file names to ensure the correct sequence
images.sort()

# Define the video writer
output_video = cv2.VideoWriter('output.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 30, (1920, 1080))

# Iterate through each image, resize it, and write it to the video
for image_name in images:
    img_path = os.path.join(image_folder, image_name)
    img = cv2.imread(img_path)
    output_video.write(img)

# Release the video writer
output_video.release()

# Load the video file using moviepy
video = VideoFileClip('output.mp4')

# Load the audio file generated by pyttsx3
audio = AudioFileClip(output_file)

# Set the duration of the video to match the duration of the audio
video = video.set_duration(audio_duration)

# Set the audio of the video to the generated voice
video = video.set_audio(audio)

# Export the final video
video.write_videofile('final_output.mp4', codec='libx264', audio_codec='aac')

# Remove the temporary files
os.remove(output_file)
os.remove('output.mp4')

print("Video created successfully!")
```

```python
import cv2
import os
import pyttsx3
import imageio
from moviepy.editor import AudioFileClip

# Prompt the user to enter the text
text = input("Enter the text to convert to speech: ")

# Initialize pyttsx3 and generate the voice
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)  # Selecting the first index for a male voice
output_file = "output.mp3"
engine.save_to_file(text, output_file)
engine.runAndWait()

# Get the duration of the generated audio
audio_duration = AudioFileClip(output_file).duration

# Path to the folder containing the images
image_folder = './images - Copy'

# Retrieve the list of image file names in the folder with supported extensions
supported_extensions = ['.jpg', '.jpeg', '.png']
images = [img for img in os.listdir(image_folder) if os.path.splitext(img)[1].lower() in supported_extensions]

# Sort the image file names to ensure the correct sequence
images.sort()

# Initialize the video writer
output_video = cv2.VideoWriter('output.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 30, (1920, 1080))

# Iterate through each image, resize it, and write it to the video
for image_name in images:
    img_path = os.path.join(image_folder, image_name)
    img = cv2.imread(img_path)
    output_video.write(img)

# Release the video writer
output_video.release()

# Load the audio file generated by pyttsx3
audio = AudioFileClip(output_file)

# Set the duration of the audio to match the duration of the video
audio = audio.set_duration(audio_duration)

# Create a video with audio using imageio
video = imageio.imread('output.mp4', plugin='imageio_ffmpeg')
video = video.set_audio(audio)

# Export the final video
video.write_videofile('final_output.mp4', codec='libx264', audio_codec='aac', fps=30)  # Replace 30 with the desired frame rate

# Remove the temporary files
os.remove(output_file)
os.remove('output.mp4')

print("Video created successfully!")
```
